# Multiple linear regression

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

## Load the data

```python
data = pd.read_csv('1.02. Multiple linear regression.csv')
data.head()
```

```python
data.describe()
```

## Create the multiple linear regression

### Declare the dependent and independent variables

```python
x = data[['SAT','Rand 1,2,3']]
y = data['GPA']
```

### Regression itself

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Calculating R-squared

```python
# rsquared value not adjusted
reg.score(x, y)
```

### Formula for Adjusted R^2

$R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$

```python
x.shape
```

### python function

```python
rsq = reg.score(x, y)
(n, p) = x.shape
def adjusted_rsq(rsq, n, p):
    return 1 - (1 - rsq) * ((n - 1) / (n - p - 1))
```

```python
adjusted_rsq(rsq, n, p) 
```

Conclusion is that one or 2 variables is unneeded

## Feature Selection
this helps us determine which values we should use. P-values help us determine what to use.

F-regression.
1. Gpa <- SAT
2. GPA <- Rand 1,2,3

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1]
p_values
```

```python
p_values.round(3)
```

## Fregression is useful for simple models

but not for more complicated situations. 
Here the p_value tells us that the second variable which is Rand variable is not vary significant (p > .2) so we should ignore it.

## Creating a summary table

```python
reg_summary = pd.DataFrame(data = x.columns.values, columns=['Features'])
reg_summary['Coefficients'] = reg.coef_
reg_summary['p-values'] = p_values.round(3)
reg_summary
```

p values tell us if variable is redundant but don't tell us how useful a varaible is.
